```python
baby_names = pd.read_csv('C:/Users/Dell/Desktop/td_python/pandas/6_Stats/data/US_Baby_Names.csv', sep=',')
baby_names.head()
```

### 4. Afficher les 10 premières lignes

```python
baby_names.head(10)
```

### 5. Supprimer les colonnes 'Unnamed: 0' et 'Id'

```python
baby_names = baby_names.drop(['Unnamed: 0', 'Id'], axis=1)
```

### 6. Y a t-il une majorité de noms masculins ou feminins dans ce jeu de données ?

```python
baby_names.groupby(['Gender'])['Gender'].count().sort_values(ascending=False).head(1)
#Si on veut connaitre les deux résultats : 
#baby_names.groupby(['Gender'])['Gender'].count()
```

### 7. Grouper les données par nom et stocker le résultats dans une variable *noms*.

```python
noms = baby_names.groupby(['Name'])
noms.head()
```

### 8. Combien de noms différents y a t-il dans ce jeu de données ?

```python
baby_names.Name.nunique()
```

### 9. Quel est le nom le plus fréquent ?

```python
#baby_names['Name'].value_counts().head(1)
baby_names['Name'].value_counts().idxmax()
```

### 10. Combien de noms différents ont la plus basse occurence ?

```python
A=baby_names['Name'].value_counts().sort_values(ascending=True)
A.value_counts().head(1)
```

### 11. Quelle est l'occurence médiane ?

```python
baby_names['Name'].value_counts().median()
```

### 12. Quele est l'écart-type des noms ?

```python
baby_names['Name'].value_counts().std(ddof = 0)
```

### 13. Afficher un résumé statistiques : moyenne, min, max, ecart-type et quartiles.

```python
baby_names.describe()
```
