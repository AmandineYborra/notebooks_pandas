### 3. Affecter les données à une variable users.

```python
users = pd.read_table("C:/Users/Dell/Desktop/td_python/pandas/Importer_&_Explorer/data/u.user", sep='|')
users.head()
```

### 4. Quel est l'age moyen par profession ?

```python
users.groupby('occupation').age.mean()
```

### 5. Quel est la proportion d'hommes par profession. Trier le résultat par ordre décroissant.

```python
users.loc[users['gender']=="M",:].groupby(['gender','occupation'])['gender'].count().sort_values(ascending=False)
```

### 6. Quels sont les amplitudes d'ages de chaque profession ?

```python
A= users.groupby('occupation').age.agg(['min', 'max'])
A['amplitudes']= A['max']-A['min']
A
```

### 7. Quels sont les ages moyens par profession et par sexe ?

```python
users.groupby(['occupation', 'gender']).age.mean()
```

### 8.  Quels sont, par profession, les pourcentages d'hommes et de femmes ?

```python
pd.crosstab(users['gender'],users['occupation']).apply(lambda x: x/x.sum()*100, axis=1)
```
